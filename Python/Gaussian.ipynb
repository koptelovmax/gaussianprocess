# Gaussian distribution

```python
import numpy

mju = 0 # mean
sigma = 0.1 # standard deviation
s = numpy.random.normal(mju, sigma, 1000)
```

```python
print(s)
```

Verify the mean and the variance

```python
abs(mju - s.mean()) < 0.01
```

```python
abs(sigma - s.std()) < 0.01
```

Display the histogram

```python
import matplotlib.pyplot

% matplotlib inline

matplotlib.pyplot.hist(s, 30)
```

```python
count,x,ignored = matplotlib.pyplot.hist(s, 30, normed=True) # First element of return is normalized to form probability density
y = 1/(sigma*numpy.sqrt(2*numpy.pi))*numpy.exp(-(x-mju)**2/(2*sigma**2))
matplotlib.pyplot.plot(x, y, linewidth=2, color='r')
matplotlib.pyplot.show()
```

## Joint distribution

To plot joint distribution **seaborn** library may be used. It is based on **pandas** which is part of Anaconda.

To use seaborn it is necessary to import pandas, then to install and import seaborn.

To install seaborn in Windows the best way is to use **conda** whis is utility that can be run from cmd. Type *conda install seaborn* in cmd to install seaborn.

```python
import pandas
import seaborn

N = 1000 # number of values
mju1 = 0
mju2 = 0
sigma1 = 0.1
sigma2 = 1

f1 = numpy.random.normal(mju1, sigma1, N)
f2 = numpy.random.normal(mju2, sigma2, N)

seaborn.jointplot(f1, f2)
seaborn.jointplot(f1, f2, kind="kde")
```
